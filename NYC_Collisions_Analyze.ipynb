```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as pk
```

```python
df = pd.read_csv("/content/NYC_Collisions.csv")

df.head(10)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'])

df['Datetime'] = df['Date'] + pd.to_timedelta(df['Time'].dt.strftime('%H:%M:%S'))

df.drop(columns=['Date', 'Time'], inplace=True)

cols = df.columns.tolist()
cols = cols[:1] + [cols[-1]] + cols[1:-1]
df = df[cols]


df.head(10)
```

```python
plt.figure(figsize=(10, 6))
df.resample('W', on='Datetime').size().plot()
plt.title('Tren Kecelakaan Mingguan')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kecelakaan')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Hour'] = df['Datetime'].dt.hour
sns.histplot(df['Hour'], bins=24, kde=False)
plt.title('Distribusi Kecelakaan Berdasarkan Jam')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Jumlah Kecelakaan')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Contributing Factor'].value_counts().head(10).plot(kind='bar')
plt.title('Faktor Penyebab Kecelakaan (Top 10)')
plt.xlabel('Faktor Penyebab')
plt.ylabel('Jumlah Kecelakaan')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df['Borough'].value_counts().plot(kind='bar')
plt.title('Jumlah Kecelakaan di Setiap Borough')
plt.xlabel('Borough')
plt.ylabel('Jumlah Kecelakaan')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x=df['Datetime'].dt.weekday, hue='Borough')
plt.title('Tren Kecelakaan Mingguan Berdasarkan Borough')
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Jumlah Kecelakaan')
plt.legend(title='Borough', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(ticks=range(7), labels=['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'])
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Persons Injured', hue='Borough')
plt.title('Kecelakaan Fatal vs. Non-fatal Berdasarkan Borough')
plt.xlabel('Jumlah Orang yang Tewas')
plt.ylabel('Jumlah Kecelakaan')
plt.legend(title='Borough', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y')
plt.show()
```

```python
pedestrian_bike_accidents = df[(df['Pedestrians Injured'] > 0) | (df['Pedestrians Killed'] > 0) | (df['Cyclists Injured'] > 0) | (df['Cyclists Killed'] > 0)]

# Visualisasi 1: Distribusi Kecelakaan yang Melibatkan Pejalan Kaki
plt.figure(figsize=(10, 6))
sns.countplot(data=pedestrian_bike_accidents, x='Borough')
plt.title('Distribusi Kecelakaan yang Melibatkan Pejalan Kaki Berdasarkan Borough')
plt.xlabel('Borough')
plt.ylabel('Jumlah Kecelakaan')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Visualisasi 2: Distribusi Kecelakaan yang Melibatkan Pengendara Sepeda
plt.figure(figsize=(10, 6))
sns.countplot(data=pedestrian_bike_accidents, x='Hour')
plt.title('Distribusi Kecelakaan yang Melibatkan Pengendara Sepeda Berdasarkan Jam')
plt.xlabel('Jam dalam Sehari')
plt.ylabel('Jumlah Kecelakaan')
plt.grid(axis='y')
plt.show()
```

### ANALYSIS BY BOROUGH "BROOKLYN"

```python
collisions_by_borough = df['Borough'].value_counts()

plt.figure(figsize=(10, 6))
collisions_by_borough.plot(kind='barh', color='skyblue')
plt.title('Total Collisions by Borough')
plt.xlabel('Total Collisions')
plt.ylabel('Borough')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df_brooklyn = df[df['Borough'] == 'Brooklyn']
contributing_factors = df_brooklyn['Contributing Factor'].value_counts().head(10)
contributing_factors.plot(kind='barh', color='skyblue')
plt.title('Top Contributing Factors to Collisions in Brooklyn')
plt.xlabel('Number of Collisions')
plt.ylabel('Contributing Factors')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
vehicle_types = df_brooklyn['Vehicle Type'].value_counts().head(10)
vehicle_types.plot(kind='barh', color='skyblue')
plt.title('Top Vehicle Types Involved in Collisions in Brooklyn')
plt.xlabel('Number of Collisions')
plt.ylabel('Vehicle Types')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_brooklyn['Datetime'] = pd.to_datetime(df_brooklyn['Datetime'])
df_brooklyn['YearMonth'] = df_brooklyn['Datetime'].dt.to_period('M')
collisions_by_month = df_brooklyn['YearMonth'].value_counts().sort_index()
collisions_by_month.plot(kind='line', marker='o', color='skyblue')
plt.title('Trend of Collisions Over Time in Brooklyn')
plt.xlabel('Year-Month')
plt.ylabel('Number of Collisions')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
df_brooklyn[['Latitude', 'Longitude']].plot(kind='scatter', x='Longitude', y='Latitude', color='skyblue', alpha=0.6, s=5)
plt.title('Locations of Collisions in Brooklyn')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
consequences = df_brooklyn[['Pedestrians Injured', 'Pedestrians Killed', 'Cyclists Injured', 'Cyclists Killed', 'Motorists Injured', 'Motorists Killed']].sum()
consequences.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen', 'darkred', 'orange', 'purple'])
plt.title('Consequences of Collisions in Brooklyn')
plt.xlabel('Consequences')
plt.ylabel('Number of People')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_brooklyn = df[df['Borough'] == 'Brooklyn']

motorists_injured_by_factor = df_brooklyn.groupby('Contributing Factor')['Motorists Injured'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
motorists_injured_by_factor.plot(kind='bar', color='skyblue')
plt.title('Top Contributing Factors to Motorists Injured in Brooklyn')
plt.xlabel('Contributing Factors')
plt.ylabel('Number of Motorists Injured')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_brooklyn_motorists_injured = df[(df['Borough'] == 'Brooklyn') & (df['Motorists Injured'] > 0)]

collisions_by_hour = df_brooklyn_motorists_injured['Datetime'].dt.hour

plt.figure(figsize=(10, 6))
collisions_by_hour.hist(bins=24, color='skyblue', edgecolor='black')
plt.title('Distribution of Collisions by Hour in Brooklyn with Motorists Injured')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Collisions')
plt.xticks(range(0, 24))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Street with highest people injuried

```python
street_count = df.groupby(['Street Name'])['Street Name'].count().reset_index(name='Collisions')
street_count = street_count.sort_values('Collisions', ascending = False).head(10)
fig, ax = plt.subplots(1, figsize=(7, 6))
street_count.plot(x = 'Street Name',
                      y = 'Collisions', kind = 'bar', color=['black', 'red', 'green', 'blue', 'cyan', 'yellow', 'lightgreen', 'pink', 'orange', 'darkred'],ax = ax ,  stacked = True)

ax.set_title('Number of People Injured Street Wise', size = 12)
ax.set_xlabel('Street', size = 12)
ax.set_ylabel('Number of People Injured', size = 12)
ax.get_legend().remove()
plt.show()
```

```python
df = df.dropna(subset=["Contributing Factor","Borough","Street Name"])
df = df[df["Contributing Factor"] != "Unspecified"]
df['Cross Street'] = df['Cross Street'].fillna('')

df.head(10)
```

```python
import geopandas as gpd

nyc_boroughs = gpd.read_file(gpd.datasets.get_path('nybb'))

nyc_boroughs = nyc_boroughs[nyc_boroughs['BoroName'].isin(['Manhattan', 'Bronx', 'Brooklyn', 'Queens', 'Staten Island'])]

accidents_by_borough = df['Borough'].value_counts().reset_index()
accidents_by_borough.columns = ['Borough', 'Accident Count']

merged_data = nyc_boroughs.merge(accidents_by_borough, left_on='BoroName', right_on='Borough')

fig, ax = plt.subplots(1, 1, figsize=(10, 6))
merged_data.plot(column='Accident Count', cmap='YlGnBu', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

for idx, row in merged_data.iterrows():
    plt.annotate(text=row['BoroName'], xy=(row['geometry'].centroid.x, row['geometry'].centroid.y), xytext=(3, 3),
                 textcoords="offset points", fontsize=8, color='black')

ax.set_title('Jumlah Kecelakaan Berdasarkan Borough')
plt.show()
```
